```python
import cv2
import matplotlib.pyplot as plt

# ===== USER SETTINGS =====
input_path = "./test_img/test_range/43.png"
output_path = "./test_img/test_range/output/43.png"

# crop region (pixel coordinates)
x1, y1 = 200, 600     # top-left
x2, y2 = 1000, 1000    # bottom-right
# =========================


# Load image
img = cv2.imread(input_path)
if img is None:
    raise ValueError("Could not load image")

# Convert BGR -> RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Original size
h, w = img.shape[:2]
print(f"Original size: width={w}, height={h}")

# Clamp crop region to image bounds
x1 = max(0, min(x1, w))
x2 = max(0, min(x2, w))
y1 = max(0, min(y1, h))
y2 = max(0, min(y2, h))

if x2 <= x1 or y2 <= y1:
    raise ValueError("Invalid crop region")

# Crop
cropped = img[y1:y2, x1:x2]
cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)

# Cropped size
ch, cw = cropped.shape[:2]
print(f"Cropped size: width={cw}, height={ch}")

# Save
cv2.imwrite(output_path, cropped)
print(f"Saved to: {output_path}")

# Show before / after
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Cropped")
plt.imshow(cropped_rgb)
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np

# ===== USER SETTINGS =====
input_path = "./test_img/test_range/output/43.png"
# =========================


# Load image
img = cv2.imread(input_path)
if img is None:
    raise ValueError("Could not load image")

# Convert color spaces
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# Split channels
h, s, v = cv2.split(hsv)
l, a, b = cv2.split(lab)

print("HSV ranges:")
print(f"H: {h.min()} → {h.max()}")
print(f"S: {s.min()} → {s.max()}")
print(f"V: {v.min()} → {v.max()}")

print()
print("LAB ranges:")
print(f"L: {l.min()} → {l.max()}")
print(f"A: {a.min()} → {a.max()}")
print(f"B: {b.min()} → {b.max()}")
```
